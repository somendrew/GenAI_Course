# Bag of Words

```python
import numpy as np
import pandas as pd
```

```python
corpus = pd.DataFrame({"text":["people watch dswithbappy",
                         "dswithbappy watch dswithbappy",
                         "people write comment",
                          "dswithbappy write comment"],"output":[1,1,0,0]})

corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
bow = cv.fit_transform(corpus['text'])
```

```python
print(cv.vocabulary_)
```

```python
bow.toarray()
```

```python
print(bow[0].toarray()),
print(bow[1].toarray())
```

```python
cv.transform(['Ring Ring Ringa, Ring Ring Ringa, Ringa Ringa Ringa Ri']).toarray()
```

```python
cv.transform(['People like to watch TV']).toarray()
```

# N Grams


**uniGram**:    `ngram_range = (1,1) `

**biGram**:  `ngram_range = (2,2)`

**triGram**:   `ngram_range = (3,3)`





```python
df = pd.DataFrame({"text":["people watch dswithbappy",
                         "dswithbappy watch dswithbappy",
                         "people write comment",
                          "dswithbappy write comment"],"output":[1,1,0,0]})

df
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(ngram_range = (2,2))
bow = cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

# TF-IDF (Term frequency- Inverse document frequency)


```python
df = pd.DataFrame({"text":["people watch dswithbappy",
                         "dswithbappy watch dswithbappy",
                         "people write comment",
                          "dswithbappy write comment"],"output":[1,1,0,0]})

df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
bow = tfidf.fit_transform(df['text'])
```

```python
print(tfidf.vocabulary_)
```

```python
bow.toarray()
```

```python
print(tfidf.idf_)
```
