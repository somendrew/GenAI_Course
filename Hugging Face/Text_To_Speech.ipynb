```python
import transformers
from transformers import pipeline
```

```python
text = "अरे तेरी माँ का भोसड़ा फट जायेगा"

model_path = "onecxi/mms-hindi-female-indic"

pipe = pipeline("text-to-audio", model_path)

output = pipe(text)
```
